```python
games['Positive'] = games['Positive'].fillna(0) / (games['Positive'].fillna(0) + games['Negative'].fillna(0))
games['Negative'] = games['Negative'].fillna(0) / (games['Positive'].fillna(0) + games['Negative'].fillna(0))
```

```python
games.to_csv('data/games.csv')
```

```python
def calculate_owners_midpoint(estimated_owners):
    if pd.isna(estimated_owners):
        return pd.NA

    owners_range = str(estimated_owners).split(' - ')
    if len(owners_range) != 2:
        return pd.NA

    return (int(owners_range[0]) + int(owners_range[1])) // 2
```

```python
games['Owners'] = games['Estimated owners'].apply(calculate_owners_midpoint)
games['Owners end range'] = games['Estimated owners'].str.split(' - ').str[1].astype(int)
games['Owners start range'] = games['Estimated owners'].str.split(' - ').str[0].astype(int)
games.drop(columns=['Estimated owners'], inplace=True)
games['Player engagement score'] = games['Peak CCU'] / games['Owners end range'].fillna(0)
```
